```python
Genesis1List = []
for genesis in range(1,32):
    Genesis1ListDict = {}
    passageCollection = 'Genesis1:'+str(genesis)
    Genesis1ListDict['passageCollection']=passageCollection
    link = 'http://api.biblia.com/v1/bible/content/kjv.text?passage='+passageCollection+'&style=bibleTextOnly&key=78107f435755fdf76ac5f8d991664d79&header=[FullVerseRef]'
    Genesis1ListDict['link']=link
    result = requests.get(link)
    Genesis1ListDict['realpassage']=result.content
    Genesis1List.append(Genesis1ListDict)
Genesis1DataFrame = pd.DataFrame(Genesis1List)
```
